```python
import numpy as np
import matplotlib.pylab as plt
```

```python
x = np.array([-1.0, 1.0, 2.0])
print(x)
# 배열 x에 들어있는 값들에 대해 조건을 부합하면 True, 아니면 False를 담은 배열을 y에 저장
y = x > 0
print(y)
# 배열 y의 자료형을 astype()을 통해 bool에서 int로 변환
y = y.astype(np.int_) # numpy 1.20.0부터 np.int 대신 np.int_ 를 사용한다고 합니다.
print(y)
```

```python
def step_function(x):
  return np.array(x > 0, dtype=np.int)

x = np.arange(-5.0, 5.0, 0.1)
y = step_function(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.show()
```

```python
def sigmoid(x):
  return 1 / (1 + np.exp(-x))
x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.show()
```

```python
def relu(x):
  return np.maximum(0, x)
x = np.arange(-5.0, 5.0, 0.1)
y = relu(x)
plt.plot(x, y)
plt.ylim(-1, 6)
plt.show()
```
